```python
# Function to remove outliers using IQR method
def remove_outliers_iqr(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Filter out the outliers
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data
```

```python
# Apply the function
df_clean = remove_outliers_iqr(df_encoded, numeric_cols)
```

```python
# Check the shape before and after
print("Original shape:", df_encoded.shape)
print("After outlier removal:", df_clean.shape)
```

**DBSCAN Clustering**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
```

```python
# select the features
cols_to_drop = ['ID', 'Response'] if 'Response' in df.columns else ['ID']
df_num = df.drop(columns=[col for col in cols_to_drop if col in df.columns], errors='ignore')
```

```python
#Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_num)
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)
```

```python
# Add cluster labels to the original DataFrame
df['DBSCAN_Cluster'] = dbscan_labels
```

```python
# Count number of points per cluster
print(df['DBSCAN_Cluster'].value_counts())
```

```python
# Visualize clusters using PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
df['PCA1'] = pca_result[:, 0]
df['PCA2'] = pca_result[:, 1]
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='DBSCAN_Cluster', palette='tab10', s=60)
plt.title('DBSCAN Customer Clustering (PCA Reduced)')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
# Only consider points not marked as noise (-1)
mask = dbscan_labels != -1
silhouette = silhouette_score(scaled_data[mask],dbscan_labels[mask])
print(f"Silhouette Score: {silhouette:.2f}")
```

```python
from sklearn.metrics import davies_bouldin_score
db_index = davies_bouldin_score(scaled_data[mask], dbscan_labels[mask])
print(f"Davies-Bouldin Index: {db_index:.2f}")
```

```python
from sklearn.metrics import calinski_harabasz_score
ch_score = calinski_harabasz_score(scaled_data[mask], dbscan_labels[mask])
print(f"Calinski-Harabasz Score: {ch_score:.2f}")
```

**Agglomerative Clustering**

```python
# Agglomerative Clustering
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_data
```

```python
# Plot the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Cluster', palette='Set2')
plt.title('Agglomerative Clustering (3 Clusters) - PCA Projection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
sil_score = silhouette_score(scaled_data, clusters)
print(f"Silhouette Score: {sil_score:.3f}")

db_score = davies_bouldin_score(scaled_data, clusters)
print(f"Davies-Bouldin Index: {db_score:.3f}")

ch_score = calinski_harabasz_score(scaled_data, clusters)
print(f"Calinski-Harabasz Index: {ch_score:.3f}")
```

**K-means Ckustering**

```python
# K-means Clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
df["Cluster"] = kmeans.fit_predict(scaled_data)
```

```python
# Use Elbow Method to determine optimal number of clusters
inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the elbow curve
plt.figure(figsize=(8, 4))
sns.lineplot(x=list(K_range), y=inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
# Fit final KMeans with optimal k (e.g., from elbow plot, assume k=4)
optimal_k = 4
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans_final.fit_predict(scaled_data)
```

```python
# View cluster summary
cluster_summary = df.groupby('Cluster').mean()
print(cluster_summary)
```

```python
# Visualize clusters 2D using PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
df['PCA1'] = pca_result[:, 0]
df['PCA2'] = pca_result[:, 1]
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set2', s=60)
plt.title('Customer Segmentation by Clusters')
plt.grid(True)
plt.show()
```

```python
pca = PCA(n_components=3)
df_pca = pca.fit_transform(scaled_data)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_pca[:, 0], df_pca[:, 1], df_pca[:, 2], c=df["Cluster"], cmap='viridis')
ax.set_xlabel("PCA Component 1")
ax.set_ylabel("PCA Component 2")
ax.set_zlabel("PCA Component 3")
plt.title("3D PCA Visualization of Clusters")
plt.show()
```

```python
# Evaluate clustering performance
silhouette_avg = silhouette_score(scaled_data, df["Cluster"])
print("Silhouette Score:", silhouette_avg)
```

**Gaussian Mixture Model**

```python
from sklearn.mixture import GaussianMixture
```

```python
# Train Gaussian Mixture Model
gmm = GaussianMixture(n_components=4, covariance_type="full", random_state=42)
gmm.fit(scaled_data)
```

```python
# Assign cluster labels
df['Cluster_GMM'] = gmm.predict(scaled_data)
```

```python
# Evaluate Clustering Performance
ilhouette_avg = silhouette_score(scaled_data, df['Cluster_GMM'])
print(f" Silhouette Score: {silhouette_avg:.2f}")
```

```python
print("Log-Likelihood Score:", gmm.score(scaled_data))
print("BIC:", gmm.bic(scaled_data))
print("AIC:", gmm.aic(scaled_data))
```
